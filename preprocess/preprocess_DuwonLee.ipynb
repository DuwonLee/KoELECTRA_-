## 1. 당황, 불안 데이터 불러오기

```python
당황_부부 = []
for i in range(1,100):
  data_name = f'Empathy_당황_부부_{i:d}'
  file_path = f'/gdrive/MyDrive/NLP 공감형 대화 데이터셋/공감형 대화(부부, 연인)/VS_당황_부부/{data_name}.tsv'

  df = pd.read_csv(file_path, delimiter='\t')

  당황_부부.append(df)

당황_부부_combined = pd.concat(당황_부부, ignore_index=True)
```

```python
당황_연인 = []
for i in range(1,74):
  data_name = f'Empathy_당황_연인_{i:d}'
  file_path = f'/gdrive/MyDrive/NLP 공감형 대화 데이터셋/공감형 대화(부부, 연인)/VS_당황_연인/{data_name}.tsv'

  df = pd.read_csv(file_path, delimiter='\t')

  당황_연인.append(df)

당황_연인_combined = pd.concat(당황_연인, ignore_index=True)
```

```python
불안_부부 = []
for i in range(1,53):
  data_name = f'Empathy_불안_부부_{i:d}'
  file_path = f'/gdrive/MyDrive/NLP 공감형 대화 데이터셋/공감형 대화(부부, 연인)/VS_불안_부부/{data_name}.tsv'

  df = pd.read_csv(file_path, delimiter='\t')

  불안_부부.append(df)

불안_부부_combined = pd.concat(불안_부부, ignore_index=True)
```

```python
불안_연인 = []
for i in range(1,46):
  data_name = f'Empathy_불안_연인_{i:d}'
  file_path = f'/gdrive/MyDrive/NLP 공감형 대화 데이터셋/공감형 대화(부부, 연인)/VS_불안_연인/{data_name}.tsv'

  df = pd.read_csv(file_path, delimiter='\t')

  불안_연인.append(df)

불안_연인_combined = pd.concat(불안_연인, ignore_index=True)
```

## 2. 당황 데이터셋 병합

```python
embarrassment_df = pd.concat([당황_부부_combined, 당황_연인_combined], ignore_index=True)
```

```python
embarrassment_df
```

### 1) 발화자 나누기

```python
embarrassment_text_to = embarrassment_df[embarrassment_df['utterance_type'] == 0]
embarrassment_text_me = embarrassment_df[embarrassment_df['utterance_type'] == 1]

embarrassment_text_to.to_csv('embarrassment_text_to', index = False)
embarrassment_text_me.to_csv('embarrassment_text_me', index = False)
```

### 2) 문장 대응시키기

```python
embarrassment_text_to_0 = embarrassment_text_to[embarrassment_text_to['utterance_type'] == 0][['utterance_text']]
embarrassment_text_me_1 = embarrassment_text_me[embarrassment_text_me['utterance_type'] == 1][['utterance_text']]

emb_merged_text = pd.concat([embarrassment_text_to_0.reset_index(drop = True), embarrassment_text_me_1.reset_index(drop = True)], axis=1)
emb_merged_text.columns = ['0', '1']

df_emb = emb_merged_text
df_emb['sentiment'] = '당황'
```
